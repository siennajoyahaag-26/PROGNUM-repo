```python
#4.4
import numpy as np
import scipy
from matplotlib import pyplot as plt
h=np.random.normal(185, 8, 100000)
plt.hist(h, bins=40)
plt.axvline(np.mean(h), color="b", label= f"The mean is: {round(np.mean(h),2)}")
plt.axvspan(185-np.std(h), 185+ np.std(h), color="r", alpha = 0.5, label = f"The standard deviation is: {round(np.std(h),2)}")
plt.legend()
plt.show()

height= eval(input("Enter the height: "))
p1 = scipy.stats.norm.cdf(height, loc=185, scale=8)
print(f"The probability of any values being smaller than and equal to the indicated height is {p1}.")
p2= scipy.stats.norm.cdf(185+8, loc=185, scale = 8)-scipy.stats.norm.cdf(185-8, loc=185, scale = 8)
print(f"The probaility of a height being located within 1 standard deviation of the height is {p2}.")
```

```python
# 4.12
from matplotlib import pyplot as plt
import numpy as np
x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

a = np.radians(30)
rotor = np.cos(a) + 1j*np.sin(a)
comp = (x+y*1j)
rotated = comp*rotor
xc= np.real(rotated)
yc= np.imag(rotated)
plt.plot(x,y)
plt.plot(xc,yc)
plt.show()
```

```python
# 4.14
import numpy as np
def funcfilter(func, seq):
    seq2 = []
    for i in range(len(seq)):
        if seq[i] > 0:
            seq2.append(seq[i])
    seqq = np.asarray(seq2)
    return func(seqq)

sq = np.array([-100, -10, 0, 10, 100])
result = funcfilter(np.log10, sq)
print("Result log10:", result)
```

```python
# 4.16
from astropy.cosmology import Planck18
import math as m

H0 = Planck18.H0.value/978
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0
z=2

def h(z):
    h= H0*m.sqrt(Om0*(1+z)**3+Ode0)
    return h
def t(z):
    t1=1/((1+z)*h(z))
    return t1
gyr = scipy.integrate.quad(t,0,z)
print(f" The lookback time for z=2, and it's uncertainty is {gyr}.")
```

```python
# 4.21
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

X = np.random.normal(0, 1, 100000) # Generating 100000 random samples from normal distribution with mean 0 and standard deviation 1
Y = np.random.normal(2, 1, 100000) # Generating 100000 random samples from normal distribution with mean 2 and standard deviation 1

plt.hist(X, bins=30, density=True, color = 'royalblue', alpha=0.8, label='X', edgecolor='black', linewidth=1.0)   # First histogram with 30 divisions, total area 1, blue, transparent, and the blacck outline 
plt.hist(Y, bins=30, density=True, color = 'lightcoral', alpha=0.8, label='Y', edgecolor='black', linewidth=1.0)   # Second histogram, same as first but in coral

xx = np.linspace(-6, 8, 400) # Creating array for normal distributions curve
pdf_x = norm.pdf(xx, 0, 1) # Computing  normal probability density function with mean 0 and standard deviation 1 at each x‑value in xx
pdf_y = norm.pdf(xx, 2, 1) # Computing  normal probability density function with mean 2 and standard deviation 1 at each x‑value in xx
plt.plot(xx, pdf_x, 'k--', linewidth=3, label='X(PDF)') # Draws a line by connecting  points (xx[i], pdf_x[i]) across all i, with black dashed line.
plt.plot(xx, pdf_y, 'k-',  linewidth=3, label='Y(PDF)') # Same idea but with solid black line 

plt.xlabel('Value') # X-axis label
plt.ylabel('Probability density') # Y-axis label
plt.title('1D Histogram + PDF')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5) # Drawing grid lines only on y direction, dashed, semi-transparent
plt.xticks(np.arange(-6, 9, 1)) # From -6 to 8, step 1

plt.text(5, 0.15, r'$\mathcal{N}(2,1)$', 
         color='red', fontsize=14, va='center' ) 
plt.arrow(4.8, 0.15, -0.6, 0,    # x_start=4.8, y_start=0.15, dx=-0.6, dy=0
          color='red', width=0.015, head_width=0.05, head_length=0.5)
plt.text(-5, 0.15, r'$\mathcal{N}(0,1)$', 
         color='blue', fontsize=14, va='center' ) 
plt.arrow(-2.8, 0.15, 0.6, 0,    # x_start=4.8, y_start=0.15, dx=-0.6, dy=0
          color='blue', width=0.015, head_width=0.05, head_length=0.5)

plt.tight_layout() # Automatically adjusts spacing
plt.show()
```

```python
X = np.random.normal(0, 1, 100000) # Generating 100000 random samples from normal distribution with mean 0 and standard deviation 1
Y = np.random.normal(2, 1, 100000) # Generating 100000 random samples from normal distribution with mean 2 and standard deviation 1

h = plt.hist2d(X, Y, bins=30, cmap='Blues', range=[[-4.4,4.2], [-2.4,6.1]])

counts = h[0].flatten() # Turns 30x30 bins array into 1d array
max_count = np.max(counts) # Finds maximum of that array

# Calculate percentiles
p1 = 1 - 0.683  # 1σ: 68.3% inside
p2 = 1 - 0.954  # 2σ: 95.4% inside  
p3 = 1 - 0.997  # 3σ: 99.7% inside

sigma1_level = np.percentile(counts, p1*100) # Finds count threshold where 31.7% of bins have higher counts (68.3% of points inside 1σ contour)
sigma2_level = np.percentile(counts, p2*100) # Finds count threshold where 4.6% of bins have higher counts (95.4% of points inside 2σ contour)
sigma3_level = np.percentile(counts, p3*100) # Finds count threshold where 0.3% of bins have higher counts (99.7% of points inside 3σ contour)

sigma_levels = np.unique([sigma3_level, sigma2_level, sigma1_level])  # Removes duplicates and makes them increasing
if len(sigma_levels) < 3:  # If still insufficient, create spaced levels
    sigma_levels = np.linspace(max_count*0.05, max_count*0.4, 3)
# Fallback if percentiles give duplicates: [115, 460, 920]
# Guarantees exactly 3 distinct increasing levels

sigma_levels = sigma_levels[sigma_levels < max_count]  
# Final safety: remove any levels ≥ peak (invalid for contouring)

# RUGGED sigma contours
extent = [h[1][0], h[1][-1], h[2][0], h[2][-1]] # Maps 30x30 matrix to data coordinates
contour = plt.contour(h[0].T, extent=extent, levels=sigma_levels,
                      colors=['red', 'red', 'red'], linewidths=2) # h[0].T = transpose 30x30 counts to match extent row-major order
plt.clabel(contour, inline=True, fontsize=12, 
           fmt={sigma_levels[0]:'3σ', sigma_levels[1]:'2σ', sigma_levels[2]:'1σ'})

plt.grid(linestyle='--', color = 'black')
plt.colorbar(h[3])

plt.xlabel('X') # X-axis label
plt.ylabel('Y') # Y-axis label
plt.xticks([-4, -2, 0, 2, 4])
plt.title('2D Histogram + Contour')
plt.tight_layout() # Automatically adjusts spacing
plt.show()
```
