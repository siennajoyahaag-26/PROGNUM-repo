```python
# 4.3
import numpy as np
u= input("Are you ready to play rock, paper, scissors? Please enter R (rock), P (paper) or S (scissors) to begin: ") # Takes user input as their choice.
if len(u)!=1: # Checks that the input is only one letter.
    u = input("Please enter only the first letter: ") # Error message
rps=np.random.randint(1, 3+1) # Creating a random integer between 1 and 3

# Assigning either R, P, or S, depending on the random integer.
if rps==1: 
    g="R"
elif rps==2:
    g="S"
else:
    g="P"
# Comparing input and computer's generation to see who wins, and printing the result.
if g==u:
    print(f"It's a tie! The computer also chose {g}.")
elif (u=="S" and g=="R") or (u=="R" and g=="P") or (u=="P" and g=="S"):
        print(f"Computer wins! It chose {g}.")
elif (u=="R" and g=="S") or (u=="P" and g=="R") or (u=="S" and g=="P"):
        print(f"You win! The computer chose {g}.")
else:
    print("Something went wrong. Please make sure you enter either P, R or S.") # Error message in case they put a different letter
```

```python
#4.4
import numpy as np
import scipy
from matplotlib import pyplot as plt
h=np.random.normal(185, 8, 100000)
plt.hist(h, bins=40)
plt.axvline(np.mean(h), color="b", label= f"The mean is: {round(np.mean(h),2)}")
plt.axvspan(185-np.std(h), 185+ np.std(h), color="r", alpha = 0.5, label = f"The standard deviation is: {round(np.std(h),2)}")
plt.legend()
plt.show()

height= eval(input("Enter the height: "))
p1 = scipy.stats.norm.cdf(height, loc=185, scale=8)
print(f"The probability of any values being smaller than and equal to the indicated height is {p1}.")
p2= scipy.stats.norm.cdf(height+8, loc=185, scale = 8)-scipy.stats.norm.cdf(height-8, loc=185, scale = 8)
print(f"The probaility of a height being located within 1 standard deviation of the height is {p2}.")
```

```python
# 4.12
from matplotlib import pyplot as plt
import numpy as np
x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

a = np.radians(30)
rotor = np.cos(a) + 1j*np.sin(a)
comp = (x+y*1j)
rotated = comp*rotor
xc= np.real(rotated)
yc= np.imag(rotated)
plt.plot(x,y)
plt.plot(xc,yc)
plt.show()
```
