```python
# 3.2
import math
import scipy.constants as s
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
objects = {} # Creating empty dictionary for names and escape velocities
for i in range(len(masses)):
    v = math.sqrt(2*s.G*masses[i]/radii[i]) # Calculating escape velocities
    sMass = str(masses[i]) # Converting mass to string in order to limit to 2 decimal places
    print(f"The mass of {names[i]} is {sMass[0:4]},")
    print(f"and its escape velocity is {v}")
    print()
    objects[names[i]] = v # Adding velocities to dictionary corresponding with the name of the object
print(objects)
```

```python
# 3.22
import matplotlib.pyplot as plt
import math as m
x = range(-5,6,1)
x2 = np.arange(-5,6,1)

def cat(a):
    j = m.cosh(a)
    return j
y = []
for i in range(len(x)):
    y.append(cat(x[i]))
y2 = []
for i in range(len(x2)):
    y2.append(cat(x2[i]))
plt.suptitle('Catenary')
plt.grid(True)
plt.xlabel('Horizontal', fontsize=14)
plt.ylabel('Vertical')
plt.xticks(x)
plt.yticks(range(-5,77,5))
plt.plot(x2, y, label = "A classic catenary")
plt.legend()
plt.show()
```
