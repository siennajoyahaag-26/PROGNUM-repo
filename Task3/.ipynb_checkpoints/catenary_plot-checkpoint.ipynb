```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
x = range(-5,6,1)
x2 = np.arange(-5,6,1)

def cat(a):
    j = m.cosh(a)
    return j
y = []
for i in range(len(x)):
    y.append(cat(x[i]))
y2 = []
for i in range(len(x2)):
    y2.append(cat(x2[i]))
plt.suptitle('Catenary')
plt.grid(True)
plt.xlabel('Horizontal')
plt.ylabel('Vertical')
plt.xticks(x)
plt.yticks(range(-5,77,5))
plt.plot(x2, y, label = "A classic catenary")
plt.legend()
plt.show()
```
